```r
library(mvtnorm)
w1_M <- matrix(c(1,1), nrow=1)
w1_S <- matrix(c(1,0,0,1), nrow=2)
w1_pdf <- function(x) {
  dmvnorm(x, mean = w1_M, sigma = w1_S)
}

x <- c(1.8,1.8)
print(w1_pdf(x))
```

```r
w2_M <- matrix(c(3,3), nrow=1)
w2_S <- matrix(c(1,0,0,1), nrow=2)
w2_pdf <- function(x) {
  dmvnorm(x, mean = w2_M, sigma = w2_S)
}

print(w2_pdf(x))
```

```r
w1_Posterior <- (w1_pdf(x)*0.5)/((w1_pdf(x)*0.5)+(w2_pdf(x)*0.5))
print(w1_Posterior)
```

```r
w2_Posterior <- (w2_pdf(x)*0.5)/((w1_pdf(x)*0.5)+(w2_pdf(x)*0.5))
print(w2_Posterior)
```

```r
plot(df, main = "Scatter Plot of non-integer lambda", 
     xlab = "X", ylab = "Y", col = "blue")
```
